```python
import scipy.fftpack as fftp
```

```python
def pad(ar):
    l  = len(ar)
    z = zeros(l)
    a = array((ar, z))
    a = a.reshape((l*2))
    return a
```

Fourrier based correlation
===============
using: 
$${\rm corr}\left(x,y\right)\left(t\right)=\mathcal{F}^{-1}\left\{ \mathcal{F}\left[x\right]\left(\omega\right)\mathcal{F}\left[y\right]\left(-\omega\right)\right\} $$

```python
def correlate_fft(x,y):
    return fftp.ifft(fftp.fft(pad(x)) * fftp.fft(pad(y[::-1])) )
```

```python
plot(correlate_fft(x,x))
plot(correlate1(x,x),".")
```

```python
plot(correlate_fft(x,y)) 
plot(correlate1(x,y),".")
```

### Marr Hildreth Edg Detector

* Smotthe image by Gaussian filter $\implies S$
* Apply Laplacian to $S$
* Find zero crossings. 

$ S = g \star I $

where $S$ is the smoothed omage, $g$ is the gaussiain and $I$ is the original image. 

The laplacian is $\Delta^2 S = (\frac{\partial^2 }{\partial x^2 } S+ \frac{\partial^2 }{\partial y^2 }  S ) $


Laplacian of Gaussian is 

$ \Delta^2 S = \Delta^2 (g \star I ) = (\Delta^2 g ) \star I  $

$\Delta^2 g = \frac{1}{\sqrt{2 \pi} \sigma^3 } \left( 2-\frac{x^2+y^2}{\sigma^2} \right) e^{-\frac{x^2+y^2 }{2 \sigma^2}}$

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
r = np.arange(-3, 3, 0.1)
X, Y = np.meshgrid(r, r)
lap_gauss  = lambda x,y,sigma:- 1/(sqrt(2*pi)*sigma**3) * (2-(x**2+y**2)/sigma**2) * exp(-(x**2+y**2)/(2*sigma**2))
Z = lap_gauss(X,Y,1)
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.jet,
        linewidth=0, antialiased=False)
```

**optimization:**
Gaussian  mask(s) can be decomposed into 2 1D masks:

$h(x,y) = (I(x,y) \star g_1(x) ) \star g_2(y) $

For laplacian-gaussian, 

$\Delta S = (I \star g_{xx}(x)) \star g(y) + (I \star g_{yy} (y) \star g(x)) $

where $g_{xx},(g_{yy})$ are second derivative in respect to $x(y)$.

We  have 4 1D convolutions. 

```python
from scipy.signal import convolve2d
```

```python
g = lambda x, sigma : 1/(sqrt(2*pi) * sigma) * exp(-x**2/(2*sigma**2))
gxx = lambda x, sigma: 1/(sqrt(2*pi)*sigma**3) * (x**2/sigma**2-1)*exp(-x**2/(2*sigma**2))
r = array([arange(-6,7,1)])

g_mask = g(r,2.)
gxx_mask = gxx(r,2.)
```

```python

lap_gauss_im = convolve2d(convolve2d(lena,gxx_mask), g_mask.T) + convolve2d(convolve2d(lena,gxx_mask.T), g_mask)
imshow(lap_gauss_im)
```

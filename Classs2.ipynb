## Derivative: ##

$ f'(x) =\lim_{\Delta x\to 0}  \frac{f(x) - f(x - \Delta x)}{\Delta x} $

Discrete Derivative: 
$\frac{df}{dx} = \frac{f(x)-f(x-1)}{1}= = f(x)-f(x-1) = f'(x)$ -- Backward difference 

$\frac{df}{dx}  = f(x+1)-f(x) = f'(x)$ -- Forward difference

$\frac{df}{dx} = f(x+1) - f(x-1) = f'(x)$ -- Cental difference (upto $1/2$? )

## Derivatives in 2D ## 

Given $f(x,y)$ the grdient is $\nabla f(x,y) = \begin{pmatrix} \frac{\partial f}{\partial x}  \\
\frac{\partial f}{\partial y}
\end{pmatrix}  = \begin{pmatrix} f_x \\ f_y \end{pmatrix} $


Which will have size and direction. 

#### Direvative mask

We derive by aplying mask 

$f_{x}\implies\frac{1}{3}\begin{pmatrix}-1 & 0 & 1\\
-1 & 0 & 1\\
-1 & 0 & 1
\end{pmatrix}$ 


$f_{y}\implies\frac{1}{3}\begin{pmatrix}1 & 1 & 1\\
0 & 0 & 0\\
-1 & 0 & -1
\end{pmatrix}$ 

apply mask on each sub-block of an image (matrix) to have the derivative in a point. 

```python
I = array ([[10,10,20,20,20]]*5)
f_x = (1./3) * array([[-1,0,1]]*3)
f_y = f_x.T

(I,"", f_x,"", f_y)
```

```python

def apply_mask(I, f):
    res = zeros_like(I)
    for (i,j) in itertools.product(xrange(1,I.shape[0]-1),
                                   xrange(1,I.shape[1]-1)):
        #Element-wise multiplication. 
        res[i,j] = sum(I[i-1:i+2,j-1:j+2] * f) 
    return res

I_x = apply_mask(I,f_x)
I_y = apply_mask(I,f_y)

I_x, I_y
```

```python
imshow(apply_mask(ibw, f_x))
```

### Correlation ###

$ (f \otimes h)(i,j)  = \sum_k \sum_l f(k,l)h(i+k,j+l) $

where $f$ is the image and $h$ is the kernel ("mask")

### Convolution ### 
$ (f \otimes h)(i,j)  = \sum_k \sum_l f(k,l)h(i-k,j-l) $

which is the same like correlation after mirroring the kernel on the x and y axis. For symmetric kernels it's the same. 

```python

from scipy.signal import correlate, correlate2d
from scipy.signal import convolve, convolve2d
```

** Mean **  - $I = \frac{I_1 + \ldots + I_n}{n} = \frac{\sum_{i=1}^{n} I_i } {n} $

** Wheghted mean **  $I = \frac{w_1 I_1 +\ldots + w_n I_n}{n} = \frac{\sum_{i=1}^n w_i I_i}{n} $ 

Of values $I_i$ and weights $w_i$. 

Applying a gassian is like applying a weighted mean.

#### Gaussian mask

```python
def gauss_filter(size, sigma):
    gauss = lambda (x,y) : exp(-(x**2+y**2)/(2*sigma**2))
    X,Y = meshgrid(arange(-(size/2),size/2+1),arange(-(size/2),size/2+1))
    return  gauss((X,Y))
gauss_filter(3,1)
```

```python
figure(0)
imshow(noisy_image); title("with noise")
figure(1)
imshow(convolve2d(noisy_image, gauss_filter(3,1))); title("gauss, 3")
figure(2)
imshow(convolve2d(noisy_image, gauss_filter(5,1))); title("gauss, 5")
figure(3)
imshow(convolve2d(noisy_image, gauss_filter(7,1))); title("gauss, 7")
figure(4)
imshow(convolve2d(noisy_image, ones((3,3))/.9)); title("averaging")
```

### Prewitt and Sobel Edge Detector

Image $\xrightarrow[\text{smoothing in } x ]{\text{average}} $
blurred $\xrightarrow[\text{filtering in x}]{\text{derivative}}$ edges in x

$f_{x}\implies\frac{1}{3}\begin{pmatrix}-1 & 0 & 1\\
-1 & 0 & 1\\
-1 & 0 & 1
\end{pmatrix}$  do both

The same with y. 

```python
figure()
imshow(convolve2d(ibw, f_x))
figure()
imshow(convolve2d(ibw, f_y))
```

#### Prewitt

```python
Delta = sqrt((convolve2d(ibw, f_x))**2 + 
             (convolve2d(ibw, f_y))**2)
imshow(Delta<50)
figure()
Delta = sqrt((convolve2d(noisy_image, f_x))**2 + 
             (convolve2d(noisy_image, f_y))**2)
imshow(Delta<70)

```

#### Sobel

```python
sobel_Fx = array([[1,0,-1],
                  [2,0,-2],
                  [1,0,-1]])
sobel_Fy = sobel_Fx.T
Delta = sqrt((convolve2d(ibw, sobel_Fx))**2 +
             (convolve2d(ibw, sobel_Fy))**2)
imshow(Delta<100)
figure()
Delta = sqrt((convolve2d(noisy_image, sobel_Fx))**2 +
             (convolve2d(noisy_image, sobel_Fy))**2)
imshow(Delta<200)
```

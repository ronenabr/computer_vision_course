```python
def convolve1d(f,h):
    res =  zeros((f.shape[0]+h.shape[0]-1))
    pad_h = pad(h,(0,f.shape[0]-h.shape[0]),'constant', constant_values=0)
    print pad_h.shape, f.shape
    for i in xrange(res.shape[0]):
        res[i] = sum(f*roll(pad_h,-i))
    return res

f = linspace(-5,5,100)**2 + rand(100)
plot(f)
plot(convolve1d(f, array([1,1,1,1])/4.))
```

```python
def convolve2d(f,h):
    res =  zeros((f.shape[0]+h.shape[0]-1, f.shape[1]+h.shape[1]-1))
    pad_h = pad(h,((0,f.shape[0]-h.shape[0]),
                   (0,f.shape[1]-h.shape[1])),'constant', constant_values=0)
    print pad_h.shape, f.shape
    for i in xrange(res.shape[0]):
        for j in xrange(res.shape[1]):
            res[i,j] = sum(f*roll(roll(pad_h,i,axis=0),j,axis=1))
    return res

import scipy.misc
im1 = imread("ex1/lena.jpg")
imshow(convolve2d(im1[:200,:200,0], ones((3,3))))
```

```python

def convolve2d2(f,h):
    res =  zeros((f.shape[0]+h.shape[0]-1, f.shape[1]+h.shape[1]-1))
    pad_f = pad(f,(((h.shape[0]+1)/2,h.shape[0]/2),
                   ((h.shape[1]+1)/2,h.shape[1]/2)),'constant', constant_values=0)
    for i in xrange(f.shape[0]):
        for j in xrange(f.shape[1]):
            res[i,j] = sum(pad_f[i:h.shape[0]+i,j:h.shape[0]+j]*h)
    return res[:f.shape[0],:f.shape[1]]

```

```python
def gauss_filter( sigma):
    size = (3*sigma+1)
    gauss = lambda (x,y) : exp(-(x**2+y**2)/(2*sigma**2))
    X,Y = meshgrid(arange(-(size/2),size/2+1),arange(-(size/2),size/2+1))
    return  gauss((X,Y))

mask = gauss_filter(1)

imshow(convolve2d2(im1,mask))
figure()
imshow(convolve2d2(im2,mask))
figure()
imshow(convolve2d2(im3,mask))
figure()
imshow(convolve2d2(im4,mask))
figure()
imshow(convolve2d2(im5,mask))
```

```python
sobel_y = array([[-1,-2,-1],
                [0,0,0],
                [1,2,1]])
sobel_x =  sobel_y.T

def sobel(f,threshold):
    temp =  sqrt(convolve2d2(f,sobel_x)**2 + convolve2d2(f,sobel_y)**2)
    temp[temp<threshold] = 0
    return temp

imshow(sobel(im1,1)); figure();
imshow(sobel(im2,1)); figure();
imshow(sobel(im3,1)); figure();
imshow(sobel(im4,1)); figure();
imshow(sobel(im5,1)); figure();
```

```python
prewitt_y = array([[1,1,1],
                [0,0,0],
                [-1,-1,-1]])
prewitt_x =  prewitt_y.T

def prewitt(f,threshold):
    temp =  sqrt(convolve2d2(f,prewitt_y)**2 + convolve2d2(f,prewitt_y)**2)
    temp[temp<threshold] = 0
    return temp

imshow(prewitt(im1,1)); figure();
imshow(prewitt(im2,1)); figure();
imshow(prewitt(im3,1)); figure();
imshow(prewitt(im4,1)); figure();
imshow(prewitt(im5,1)); figure();
```

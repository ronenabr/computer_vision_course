```python
%matplotlib inline
from pylab import *
import itertools
gray()
from scipy.signal import convolve2d
```

```python
def gauss_filter(size, sigma):
    gauss = lambda (x,y) : exp(-(x**2+y**2)/(2*sigma**2))
    X,Y = meshgrid(arange(-(size/2),size/2+1),arange(-(size/2),size/2+1))
    return  gauss((X,Y))
```

```python
gauss_filter(3,1)
m_x  = array([[-1,0,1]]*3)  * (1/3.)
m_y = m_x.T 
ker_g = gauss_filter(5, 1.5)
def dx_image(I):
    return  convolve2d(I, m_x)

def dy_image(I):
    return  convolve2d(I, m_y)

def GetHarrisCorrners(image, k=0.06):
    Ix = dx_image(image)
    Iy = dy_image(image)
    a = convolve2d(Ix**2,ker_g)
    c = convolve2d(Iy**2,ker_g)
    b2 = convolve2d(Ix*Iy,ker_g)**2
    return (1-2*k)*a*c-b2 - k*(a**2 + c**2) 

epsilon=1e-8;
def GetHarrisCorrners2(image, k=0.06):
    Ix = dx_image(image)
    Iy = dy_image(image)
    a = convolve2d(Ix**2,ker_g)
    c = convolve2d(Iy**2,ker_g)
    b2 = convolve2d(Ix*Iy,ker_g)**2
    return (a*c-b2)/(a+c+epsilon)
```

```python
def rgb2gray(rgb):

    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b

    gray = gray.astype(int)
    gray = rgb[:,:,0]
    
    return gray
```

```python
def GetHarrisTrackPoints(R,threshold=20000):
    trackPoints = []    
    window_size = 5
    count = 0;
    for i,j in itertools.product(xrange(window_size,R.shape[0]-window_size),xrange(window_size,R.shape[1]-window_size)) :
        if(R[i,j] < threshold ):
            continue
        isMaxima = True
        for k,l in itertools.product(range(-window_size, window_size+1),range(-window_size, window_size+1)):
            if (R[i+k,j+l] - R[i,j]) > 1e-3:
                isMaxima = False
                break
        if isMaxima:
            trackPoints.append((i,j))
            count += 1
    print "count", count
    return trackPoints

def GetImageTrackPoints(image, threshold=1000, k=0.06):    
    R = GetHarrisCorrners(image, k)
    imshow(R)
    colorbar()
    return GetHarrisTrackPoints(-R, threshold), R
```

```python
%matplotlib qt
figure(figsize=(15,15))
image1 = imread(r"klt/Seq1/0135.jpeg")
image2 = image1[:,:,1]
image2 = rgb2gray(image1)

trackPoints, R = GetImageTrackPoints(image2, 100000,0.005)
trackPoints = array(trackPoints).T
imshow(image2)
scatter(trackPoints[1], trackPoints[0], s=80, facecolors='none', edgecolor='r')
#figure(figsize=(15,15))
#imshow(R)
#scatter(trackPoints[1], trackPoints[0], s=80, facecolors='none', edgecolor='r')



#figure(figsize=(15,15))
#image1 = imread(r"klt/Seq1/0137.jpeg")
#image2 = image1[:,:,1]
#imshow(image2)
#trackPoints = GetImageTrackPoints(image2, 35000, 0.09)
#trackPoints = array(trackPoints).T
#scatter(trackPoints[1], trackPoints[0], s=80, facecolors='none', edgecolor='r')
```

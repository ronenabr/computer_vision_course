```python
def gauss_filter(sigma, ker_size = None):
    if not ker_size:
        ker_size = int(sigma*5)
        if (ker_size%2):
            ker_size += 1
    gauss = lambda (x,y) : exp(-(x**2+y**2)/(2*sigma**2))
    X,Y = meshgrid(arange(-(ker_size/2),ker_size/2+1),arange(-(ker_size/2),ker_size/2+1))
    return  gauss((X,Y))
```

```python
def make_odd(i):
    if i%2 == 0:
        return i+1
    return i

def computer_kerr_size(sigmas):
    sigma = max(array(sigmas))
    return make_odd(int(sigma*5))
    
```

```python


def L(I, sigma1, sigma2=None, ker_size=None) :
    if not sigma2:
        sigma2 = sqrt(2) * sigma1
    if not ker_size:
        ker_size = computer_kerr_size([sigma1, sigma2])
    
    return scipy.signal.convolve2d(I, 
            (gauss_filter(sigma2,ker_size)-gauss_filter(sigma1,ker_size))/((sqrt(2)-1)*sigma1**2))

def L2(I, sigma1, sigma2=None) :
    if not sigma2:
        sigma2 = sqrt(2) * sigma1
    ker_size = make_odd(int(sigma2*5))
    return scipy.signal.convolve2d(I, 
            gauss_filter(sigma2,ker_size))  - scipy.signal.convolve2d(I, 
            gauss_filter(sigma1,ker_size)) 
```

```python
figure(figsize=(10,10))
imshow(L(img1,3), cmap='gray')
figure(figsize=(10,10))
imshow(L2(img1,3), cmap='gray')
figure(figsize=(10,10))

imshow(L2(img1,3)-L(img1,3), cmap='gray')
colorbar()

```

```python
def is_local_max(arr,i,j,k):
    is_max = 0
    is_min = 0
    for x in range(-1,2):
        for y in range(-1,2):
            for z in range(-1,2):
                if not ((x==0) and (y==0) and (z==0)):
                    if (arr[i,j,k] > arr[i+x,j+y,k+z]):
                        is_max += 1
                    else:
                        is_min += 1
    if is_max == 8 or is_min==8:
        return True

def find_local_max(arr):
    results = []
    for i in range(1,arr.shape[0]-1):
        for j in range(1,arr.shape[1]-1):
            for k in range(1,arr.shape[2]-1):
                if is_local_max(arr,i,j,k):
                    results.append((i,j,k)) 
    return results
```
